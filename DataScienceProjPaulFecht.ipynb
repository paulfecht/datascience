```python
# inner join
newTable = pd.merge(countryCodes, GDP, on='Country') ## merging two spreadsheets or tables(same idea)

                                                              #####  CRITICAL #####
newTable.set_index('Int Olympic Committee code', inplace=True)## making the new index so that when
                                                              ## we go to make a new column in our
                                                              ## destination table it can use the key of
                                                              ## the series we created above with the key
                                                              ## of the destination table to make a match.
newTable    ## let's take a look at the result
```

```python
# 
newTable['medals'] = test # this series has as it's key the NOC country 3-letter codes
newTable
```

```python
newTableWhittled = newTable
newTableWhittled.reset_index()
```

```python
newTableWhittled['meanGDP'] = newTableWhittled.mean(axis=1)
newTableWhittled
```

```python
newTableWhittledMore = newTableWhittled[['Country', 'medals', 'meanGDP']]

newTableWhittledMore.reset_index(inplace=True)
newTableWhittledMore.dropna(axis=0, how='any') # get rid of nulls
newTableWhittledMore.sort_values(by='medals')
evenSmaller = newTableWhittledMore.nlargest(10,'medals')
evenSmaller
```
